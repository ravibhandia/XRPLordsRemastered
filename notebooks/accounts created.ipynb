```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

## Loading data

```python
data=pd.read_csv("~/Desktop/MEng/ripple/Create_Accounts_Num.csv")
data=pd.DataFrame(data)
data.head()
data.drop_duplicates()
```

## Research on Parent Account

```python
#parent freq
parent=pd.DataFrame(data['parent'].value_counts())
parent=pd.DataFrame(parent).reset_index()


#parent-account
parent_account={};
for value in parent['index']:
   son =data[data['parent']==value]['account']
   parent_account[value] = son
    
#both parent&account
acc=set(data['account'])
par=set(data['parent'])
par_acc=acc.intersection(par)
len(par_acc)
## there are 18046 accounts both parent&account

count=[];
for key,value in parent_account.items():
    value1=set(value)
    inter=value1.intersection(par_acc)
    count.append(len(inter))
## count=2st effect

parent['2st'] = count
parent.columns = ['parent','account_created','2st_account_created']

##analysis
c=filter(lambda x :x>0 ,count)
len(list(c))
#there are 5762 parents having 2st effect


##save
parent.to_csv('~/Desktop/MEng/ripple/parent.csv',index = None, header=True)
```
